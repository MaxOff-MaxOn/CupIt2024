```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv", parse_dates=["date_of", "ctl_loading_date"])
df_no_dub = df.drop_duplicates(subset=["ctl_loading", "benchmark_id"])
weak_df_no_dub = df.drop_duplicates(subset=["ctl_loading", "benchmark_id", "quote"])
df_no_dub
```

```python
df[df.duplicated(subset=["ctl_loading", "benchmark_id"], keep=False)].iloc[:20,:]
```

```python
concat_list = []
for loading in df["ctl_loading"].unique():
    quote_date = df[df["ctl_loading"] == loading]["date_of"].unique()[0]
    loading_date = df[df["ctl_loading"] == loading]["ctl_loading_date"].unique()[0]
    new_df = df[((df["date_of"] == quote_date) & (df["ctl_loading_date"] < loading_date)) | (df["ctl_loading"] == loading)]
    new_df.loc[:,"ctl_loading"] = loading
    concat_list.append(new_df)
new_concat_df = pd.concat(concat_list)
new_concat_df
    
```

```python
new_concat_df.reset_index(inplace=True)
new_concat_df.sort_values(by=["ctl_loading_date"], ascending=True)
new_concat_df
```

```python
new_concat_df.drop(columns=["index"], inplace=True)
```

```python
df_no_dublicates = new_concat_df.drop_duplicates(subset=["ctl_loading", "benchmark_id"], keep='last')
weak_df_no_dublicates = new_concat_df.drop_duplicates(subset=["ctl_loading", "benchmark_id", "quote"], keep='last')
```

```python
aggregated_quantity = new_concat_df.groupby(["ctl_loading"]).size()
aggregated_quantity
```

```python
aggregated_quantity_no_dublicates = df_no_dublicates.groupby(["ctl_loading"]).size()
aggregated_quantity_no_dublicates
```

```python
weak_aggregated_quantity_no_dublicates = weak_df_no_dublicates.groupby(["ctl_loading"]).size()
weak_aggregated_quantity_no_dublicates
```

```python
quantity = df.groupby(["ctl_loading"]).size()
quantity_no_dub = df_no_dub.groupby(["ctl_loading"]).size()
weak_quantity_no_dub = weak_df_no_dub.groupby(["ctl_loading"]).size()
```

```python
loadings = pd.concat([quantity, weak_quantity_no_dub, quantity_no_dub, aggregated_quantity, weak_aggregated_quantity_no_dublicates, aggregated_quantity_no_dublicates], axis=1).rename(columns={0: 'quan', 1: "quan_no_weak_dub", 2: 'quan_no_dub', 3: 'agg_quan', 4: "agg_quan_no_week_dub", 5: 'agg_quan_no_dub'})
loadings
```
